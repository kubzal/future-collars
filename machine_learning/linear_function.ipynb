```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
```

```python
def mean_squared_error(x, y, slope, intercept):
    """
    Calculate the Mean Squared Error (MSE) between actual y-values and predicted y-values from a linear model.

    Parameters:
    x (array-like): Independent variable values.
    y (array-like): Actual dependent variable values.
    slope (float): Slope of the linear model.
    intercept (float): Intercept of the linear model.

    Returns:
    float: The MSE value.
    """
    y_pred = slope * x + intercept
    errors = y - y_pred
    mse = np.mean(errors**2)
    return mse
```

```python
# Adjustable parameters
slope = 1  # Change this to adjust the slope
intercept = 0  # Change this to adjust the intercept

# Generate x and y values
x = np.linspace(-10, 10, 100)
y = slope * x + intercept
```

```python
# Create the plot
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(x, y, label=f"y = {slope}x + {intercept}")
ax.axhline(0, color="gray", linewidth=0.5)
ax.axvline(0, color="gray", linewidth=0.5)

# Set axis limits to focus on the first quadrant
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
# ax.set_ylim(0, slope * 10 + intercept)

# Set tick intervals to 1
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
ax.yaxis.set_major_locator(ticker.MultipleLocator(1))

# Add labels and title
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("Linear Function Plot")
ax.legend()
ax.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Set random seed for reproducibility
np.random.seed(69)
```

```python
# Generate random data points
num_points = 20
x_rand = np.random.uniform(0, 10, num_points)
noise = np.random.normal(0, 0.5, num_points)

y_rand = 0.5 * x_rand + 3 + noise  # Underlying true line: y = 0.5x + 3
```

```python
# First Plot: Only the random data points
fig1, ax1 = plt.subplots(figsize=(6, 6))
ax1.scatter(x_rand, y_rand, color="red", label="Random Data Points")
ax1.axhline(0, color="gray", linewidth=0.5)
ax1.axvline(0, color="gray", linewidth=0.5)
ax1.set_xlim(0, 10)
ax1.set_ylim(0, 10)
ax1.xaxis.set_major_locator(ticker.MultipleLocator(1))
ax1.yaxis.set_major_locator(ticker.MultipleLocator(1))
ax1.set_aspect("equal")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_title("Random Data Points")
ax1.legend()
ax1.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Second Plot: Data points with an interactive line
fig2, ax2 = plt.subplots(figsize=(6, 6))
plt.subplots_adjust(left=0.1, bottom=0.25)

# Initial parameters for the line
slope = 1
intercept = 0

x_vals = np.array([0, 10])
line_vals = slope * x_vals + intercept

[line_plot] = ax2.plot(x_vals, line_vals, color="blue", label="Fitted Line")
scatter_plot = ax2.scatter(x_rand, y_rand, color="red", label="Random Data Points")

ax2.axhline(0, color="gray", linewidth=0.5)
ax2.axvline(0, color="gray", linewidth=0.5)
ax2.set_xlim(0, 10)
ax2.set_ylim(0, 10)
ax2.xaxis.set_major_locator(ticker.MultipleLocator(1))
ax2.yaxis.set_major_locator(ticker.MultipleLocator(1))
ax2.set_aspect("equal")
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax2

# Calculate MSE
mse_value = mean_squared_error(x_rand, y_rand, slope, intercept)
print(f"Mean Squared Error: {mse_value}")
```

```python
# Assuming you want to adjust the intercept and keep the slope fixed
slope = 0.5

# Range of intercept values to evaluate
intercept_values = np.linspace(-1, 6, 100)

# Compute MSE for each intercept value
mse_values = [
    mean_squared_error(x_rand, y_rand, 0.5, intercept) for intercept in intercept_values
]

# Find the intercept with the minimum MSE
min_mse = min(mse_values)
optimal_intercept = intercept_values[mse_values.index(min_mse)]

# Plot MSE vs. Intercept
plt.figure(figsize=(8, 6))
plt.plot(intercept_values, mse_values, label="MSE vs. Intercept")
plt.axvline(
    optimal_intercept,
    color="red",
    linestyle="--",
    label=f"Optimal Intercept: {optimal_intercept:.2f}",
)
plt.xlabel("Intercept")
plt.ylabel("Mean Squared Error")
plt.title("MSE for Different Intercept Values (Slope Fixed)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
